```python
import random

def GreetUser():
    print("Welcome to GenZ Life Simulator!")
    name = input("What's your name? ")
    age = input("How old are you? ")
    user = {"name": name, "age": age}
    print(f"Hey {name}! Ready to start your life simulation?")
    return user

responses = {
    "hi": "Hey bestie!",
    "how are you": "I'm good, how about you?",
    "bye": "See you soon!"
}

def GetResponse(user_input):
    user_input = user_input.lower()
    return responses.get(user_input, "Hmm... I don't know what to say.")

def EndChat():
    print("Goodbye! See you next time.")

player_stats = {
    "energy": 100,
    "mood": "Neutral",
    "points": 0
}

def ShowStatus():
    print("\nYour current stats:")
    for key, value in player_stats.items():
        print(f"{key.capitalize()}: {value}")
    print()

def UpdateStats(activity):
    if activity == "study":
        player_stats["energy"] -= 20
        player_stats["points"] += 10
        player_stats["mood"] = "Focused"
    elif activity == "eat":
        player_stats["energy"] += 15
        player_stats["mood"] = "Satisfied"
    elif activity == "play games":
        player_stats["energy"] -= 10
        player_stats["points"] += 5
        player_stats["mood"] = "Happy"
    elif activity == "sleep":
        player_stats["energy"] += 20
        player_stats["mood"] = "Rested"
    elif activity == "workout":
        SuggestWorkout()

def EndGame():
    print("\nGame Over! Here are your final stats:")
    ShowStatus()
    print("Thanks for playing GenZ Life Simulator!")

def ActivityMenu():
    while True:
        print("\nChoose an activity:")
        print("1. Study")
        print("2. Eat")
        print("3. Play Games")
        print("4. Sleep")
        print("5. Workout (New Feature)")
        print("6. Quit")

        choice = input("Enter your choice: ").strip()

        if choice == "1":
            UpdateStats("study")
        elif choice == "2":
            UpdateStats("eat")
        elif choice == "3":
            UpdateStats("play games")
        elif choice == "4":
            UpdateStats("sleep")
        elif choice == "5":
            UpdateStats("workout")
        elif choice == "6":
            EndGame()
            break
        else:
            print("Invalid choice. Try again.")
            continue

        ShowStatus()

def SuggestWorkout():
    energy = player_stats["energy"]

    print("\nWorkout Reminder: Time to move your body!")

    if energy >= 70:
        print("High energy detected! Suggested workout:")
        print("- Full-body workout: Push-ups, squats, burpees, and lunges.")
        player_stats["energy"] -= 30
        player_stats["mood"] = "Energized"
        player_stats["points"] += 10
    elif 40 <= energy < 70:
        print("Moderate energy detected. Suggested workout:")
        print("- Upper body: Light push-ups and dumbbell curls.")
        player_stats["energy"] -= 20
        player_stats["mood"] = "Active"
        player_stats["points"] += 7
    elif 20 <= energy < 40:
        print("Low energy detected. Suggested workout:")
        print("- Stretching and light yoga.")
        player_stats["energy"] -= 10
        player_stats["mood"] = "Relaxed"
        player_stats["points"] += 4
    else:
        print("You are too tired to workout. Please rest first.")
        player_stats["mood"] = "Exhausted"

def main():
    user = GreetUser()
    while True:
        user_input = input("\nTalk to me: ")
        if user_input.lower() == "bye":
            EndChat()
            break
        elif "play" in user_input.lower() or "game" in user_input.lower():
            play = input("Do you want to play GenZ Life Simulator with me? (yes/no): ").lower()
            if play == "yes":
                ActivityMenu()
            else:
                continue
        else:
            print(GetResponse(user_input))

main()
```
